# Etapa 1: Têndencia de imigração de brasileiros para o Canadá (Matplotlib)

## Tratando Dados e Extraindo

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ThiagoFukuda/Data-Visualization-Projeto-Final/refs/heads/main/Dados/imigrantes_canada.csv')
df.head()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil.head(10)
```

## Demonstração Visual

#### Gráfico

```python
import matplotlib.pyplot as plt
```

```python
dados_brasil.plot(x='ano', y='imigrantes')
plt.title('Imigração do Brasil para o Canadá\n 1980 a 2013')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.show()
```

#### Figuras

```python
anos_selecionados = [1980, 1985, 1990, 1995, 2000, 2005, 2010]
fig, ax = plt.subplots()

ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=2, color='g')
ax.set_xticks([str(ano) for ano in anos_selecionados])
ax.set_title('Imigração do Brasil para o Canadá\n 1980 a 2013', fontsize=16, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

##### Análise gráfica em diferente formas

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_xticks([str(ano) for ano in anos_selecionados])
axs[0].set_title('Imigração do Brasil para o Canadá\n 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Imigração do Brasil para o Canadá\n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid(True, linestyle='--')

plt.show()
```

```python
#Informações do Boxplot(gráfico da direita)
dados_brasil.describe()
```

##### Comparação entre diferentes países

```python
paises = ['Brasil', 'Colômbia', 'Argentina', 'Peru']
titulos = ['Brasil', 'Colômbia', 'Argentina', 'Peru']

ymin = 0
ymax = 7000

fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.suptitle('As tendências de imigração para os países da América do Sul para o Canadá\n 1980 a 2013')

for ax, pais, titulo in zip(axs.flat, paises, titulos):
    ax.plot(df.loc[pais, anos])
    ax.set_title(titulo)
    ax.set(xlabel='Ano', ylabel='Número de imigrantes')
    ax.xaxis.set_major_locator(plt.MaxNLocator(5))
    ax.set_ylim(ymin, ymax)

plt.tight_layout()
plt.show()
```

# Etapa 2: As tendências de imigração para os países da América do Sul para o Canadá usando (Matplotlib)

## Demonstração Visual

### Gráfico Simples

```python
df
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
cores = ['#FF5733', '#33FF57', '#3357FF', '#F0E68C', '#FF33A1', '#A1FF33', '#33A1FF', '#A133FF', '#FFA500', '#8A2BE2', '#FF4500', '#2E8B57']

fig, ax = plt.subplots(figsize=(15, 5))
ax.bar(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('As tendências de imigração para os países da América do Sul para o Canadá\n1980 a 2013', fontsize=16)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

### Gráfico Detalhado

```python
america_sul_sorted = america_sul.sort_values(by='Total', ascending=True)

cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('gray')

fig, ax = plt.subplots(figsize=(15, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil quarto país com mais \n imigrantes para o Canadá de 1980 a 2013', fontsize=16)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 100, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```

## Salvando os Gráficos Informativos

```python
fig, ax = plt.subplots()

ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=2, color='g')
ax.set_xticks([str(ano) for ano in anos_selecionados])
ax.set_title('Imigração do Brasil para o Canadá\n 1980 a 2013', fontsize=16, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('imigracao_brasil_canada.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil quarto país com mais \n imigrantes para o Canadá de 1980 a 2013', fontsize=16)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 100, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

# Etapa 3: Top 10 Global de países que tiveram mais Imigrantes para o Canadá (Seaborn)

## Tratando Dados

```python
import seaborn as sns
sns.set_theme()
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

## Demonstração Visual

### Gráfico

```python
sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')

plt.show()
```

```python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')

ax.set(
       title='Top 10 países que mais migraram para \n o Canadá entre 1980 a 2013',
       xlabel='Número de Imigrantes',
       ylabel='País'
      )
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h',palette='deep')

ax.set_title('Top 10 países que mais migraram para \n o Canadá entre 1980 a 2013', fontsize=16)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')
```

### Paleta de Cores

#### Sequencial

```python
def gerar_grafico_paleta(paleta):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h',palette=paleta)
  ax.set_title('Top 10 países que mais migraram para \n o Canadá entre 1980 a 2013', fontsize=16)
  ax.set_xlabel('Número de Imigrantes', fontsize=14)
  ax.set_ylabel('')

plt.show()
```

```python
gerar_grafico_paleta('rocket')
```

```python
gerar_grafico_paleta('Blues')
```

```python
gerar_grafico_paleta('coolwarm_r')
```

#### Categórica

```python
gerar_grafico_paleta('tab10')
```

### Temas

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('deep')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('deep')
```

```python
sns.set_theme(style='white')
gerar_grafico_paleta('deep')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('deep')
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h',palette='deep')

ax.set_title('Top 10 países que mais migraram para o Canadá', loc='left', fontsize=16)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')
sns.despine()

plt.show()
```

# Etapa 4: Imigração de todos os países da América do Sul (Plotly)

### Gráfico Interativo

```python
import plotly.express as px
```

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes', title='Imigração do Brasil para o Canadá entre 1980 e 2013')

fig.update_traces(line_color='green', line_width=4)

fig.update_layout(width=1000, height=500,
                  xaxis={'tickangle':-45},
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  title_font_family='Arial',
                  title_x=0.5,
                  xaxis_title='Ano',
                  yaxis_title='Número de Imigrantes'
                  )
fig.show()
```
